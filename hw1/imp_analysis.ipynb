```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = np.load("results.npy")
```

```python
n_trees = [el[0] for el in results]
misclf = [el[1] for el in results]
sd = [el[2] for el in results]
```

```python
n_trees
```

```python
plt.plot(sd)
```

```python
plt.figure(figsize=(9, 6))
plt.errorbar(n_trees, misclf, yerr=sd, fmt='-o', ecolor="black", capsize=5, label="Misclassification Rate")
plt.legend()
plt.xlabel("Number of trees", fontsize=14)
plt.ylabel("Misclassification rate", fontsize=14)
for i, (x, y) in enumerate(zip(range(0, 501, 50), misclf)):
    plt.text(x-10, y+.008, f"{(sd[i]*100):.2f}%")

plt.xticks(range(0, 501, 50), rotation=45)
plt.grid(True)
```

```python
sd
```

```python
np.array(misclf) - np.array(sd), np.array(misclf) + np.array(sd) 
```

```python
# Alternatively, using shaded area
plt.figure(figsize=(10, 3))
plt.plot(n_trees, misclf, '-o', label="Misclassification Rate")
plt.fill_between(n_trees, 
                 np.array(misclf) - np.array(sd),
                 np.array(misclf) + np.array(sd),alpha=.2, linestyle="--")
# plt.ylim(0.1, .4)
plt.legend()
plt.xlabel("Number of trees", fontsize=14)
plt.ylabel("Misclassification rate", fontsize=14)

for i, (x, y) in enumerate(zip(range(0, 501, 50), misclf)):
    plt.text(x-10, y+.006, f"{(sd[i]*100):.2f}%")

plt.xticks(range(0, 501, 50), rotation=45)
plt.grid(True)
```
